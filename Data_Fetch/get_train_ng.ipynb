```python
import polars as pl
import os
from utils_tools.data_preprocessing.data_deal import DataProcessor

# 设置输出路径
output_path = "/data/inverter_power_data/processed_data/abnormal_data/"
output_file = os.path.join(output_path, "all_error.parquet")

# 确保输出目录存在
os.makedirs(output_path, exist_ok=True)

# 读取两个 parquet 文件
error1 = pl.read_parquet('/home/franco/power/inverter1/inverter/code_v7/data/train/ng/error_data6.parquet')
error2 = pl.read_parquet('/home/franco/power/inverter1/inverter/code_v7/data/train/ng/error_data7.parquet')

# 合并两个 DataFrame
all_error = pl.concat([error1, error2])

# 去重
all_error = all_error.unique()

# 数据处理
all_error, all_columns = DataProcessor(all_error, all_columns).main1()

# 直接赋值 sn_ym 为 sn 列的值
all_error = all_error.with_columns(all_error["sn"].alias("sn_ym"))

# 保存到 parquet 文件
all_error.write_parquet(output_file)

# 打印 sn_ym 列
print(all_error["sn_ym"])

print(f"数据已成功保存到: {output_file}")


```
