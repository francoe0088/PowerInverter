```python
import pandas as pd
import numpy as np

# 设置随机种子以确保结果可复现
RANDOM_STATE =None
np.random.seed(RANDOM_STATE)

# 1. 读取 CSV 文件
combined_all = pd.read_csv('/home/franco/power/inverter1/inverter/code_v4/combined_all.csv')
print(f"总样本数量: {len(combined_all)}")

# 2. 分离正样本和负样本
positive_samples = combined_all[combined_all['label'] == 0]
print(len(positive_samples))
negative_samples = combined_all[combined_all['label'] == 1]
print(len(negative_samples))

# 3. 计算20%样本的数量并随机抽取
positive_sample_20=positive_samples
negative_sample_20=negative_samples

# 查看抽取后的样本数量
print(f"抽取的正样本数量: {positive_sample_20.shape}")
print(f"抽取的负样本数量: {negative_sample_20.shape}")

# 4. 划分训练集和验证集
# 定义划分比例
train_frac = 0.85  # 80% 训练集
val_frac = 0.15    # 20% 验证集

def manual_train_val_split(data, train_fraction, random_state):
    """
    手动划分训练集和验证集。
    
    Parameters:
    - data (pd.DataFrame): 要划分的数据集
    - train_fraction (float): 训练集所占比例
    - random_state (int): 随机种子
    
    Returns:
    - train_data (pd.DataFrame): 训练集
    - val_data (pd.DataFrame): 验证集
    """
    # 随机打乱数据
    shuffled_data = data.sample(frac=1.0, random_state=random_state).reset_index(drop=True)
    # 计算训练集的样本数量
    train_size = int(len(shuffled_data) * train_fraction)
    # 划分训练集和验证集
    train_data = shuffled_data.iloc[:train_size].reset_index(drop=True)
    val_data = shuffled_data.iloc[train_size:].reset_index(drop=True)
    return train_data, val_data

# 对正样本进行划分
train_positive, val_positive = manual_train_val_split(
    positive_sample_20, 
    train_fraction=train_frac, 
    random_state=RANDOM_STATE
)

# 对负样本进行划分
train_negative, val_negative = manual_train_val_split(
    negative_sample_20, 
    train_fraction=train_frac, 
    random_state=RANDOM_STATE
)

# 合并训练集和验证集
train_data = pd.concat([train_positive, train_negative]).sample(frac=1.0, random_state=RANDOM_STATE).reset_index(drop=True)  # 打乱顺序
val_data = pd.concat([val_positive, val_negative]).sample(frac=1.0, random_state=RANDOM_STATE).reset_index(drop=True)        # 打乱顺序

# 查看训练集和验证集的数量
print(f"训练集数量: {train_data.shape}")
print(f"验证集数量: {val_data.shape}")

# 5. 保存到 CSV 文件
train_csv_path = '/home/franco/power/inverter1/inverter/code_v4/train_data.csv'
val_csv_path = '/home/franco/power/inverter1/inverter/code_v4/val_data.csv'

train_data.to_csv(train_csv_path, index=False)
val_data.to_csv(val_csv_path, index=False)
count_1 = (train_data['label'] == 1).sum()
print(count_1)
print(f"训练集已保存到: {train_csv_path}")
print(f"验证集已保存到: {val_csv_path}")
```
