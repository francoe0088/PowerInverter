```python
import polars as pl

# 读取 Parquet 文件
stats_ok = pl.read_parquet('/home/franco/power/inverter1/inverter/code_v4/data/data_all/20250104_final_stats_test.parquet')

# 将 'sn' 列转换为字符串类型，并创建一个新的列 'sn_str' 以保留原始 'sn' 值
stats_ok = stats_ok.with_columns(
    pl.col("sn").cast(pl.Utf8).alias("sn_str")
)

# 提取 'sn_str' 列的前6个字符，并创建新的列 'sn_prefix'
stats_ok = stats_ok.with_columns(
    pl.col("sn_str").str.slice(0, 6).alias("sn_prefix")
)

# 方法一：使用 groupby 和 count，然后按 count 降序排序
prefix_counts = stats_ok.groupby("sn_prefix").agg(
    pl.count().alias("count")
).sort("count", descending=True)  # 使用 sort_by 并指定 descending

print("每个唯一前6字符的出现次数 (使用 groupby 和 sort_by):")
print(prefix_counts)




```
